```python
#encapsulation, intro to oop

class person:
    x=10

    def __init__(self,name,age):
        self.name = name
        self.age = age

    def display(self): #pass=dummy architecture
        print("name:",self.name,",age:",self.age)
        print("person")

s=person('shahira',20)
s1 = person('khan', 22)

s.display()
s1.display()

print(s.x)
print(s1.x)
```

```python
#inheritance

class person:
    def __init__(self):
        self.x = 10

    def display(self):
        print("parent class")
        print(f"parent method, x= {self.x}")

class person2(person):
    pass

class student(person):
    def __init__(self):
        super().__init__() #for accessing parent variables
        self.y = 20

    def display(self): #overriding, priority to its own funtion of same name
        print("child class")
        #self.display()

    def display1(self):
        super().display() #calls parent function
        self.display()

s = student()
s.display1()
print(s.y)
```

```python
#multiple inheritance

class Parent1:
    def greet_parent1(self):
        print("Hello from Parent1")

    def greet_parent(self):
        print("hello 1")
        super().greet_parent()

class Parent2:
    def greet_parent2(self):
        print("Hello from Parent2")

    def greet_parent(self):
        print("hello 2")

class Child(Parent1, Parent2):
    def greet_child(self):
        super().greet_parent()
        print("Hello from Child")

c = Child()
# c.greet_parent1()
# c.greet_parent2()
c.greet_child()
```

```python
class varFunc: 

    def length(self, *s): #polymorphism overriding(prints any variable type), ** for dictionary(key, value pair)
        print(s)
    
s=varFunc()
s.length(1,2)
s.length(4,5)
s.length("shahira","khan")
s.length(1, 'ah')


class func:
    def length(self, **s):
        print(s)

x = func()
x.length(first_name = 'shahira', last_name = 'khan')

#teacher's way
class abc:
    def sumall(self,*args):
        #return sum(args)
        return args
    
a=abc()
print(a.sumall(1,2))
print(a.sumall(1,2,3))
print(a.sumall(4,5,8,323))

```

```python
#encapsulation(modes of variables)


class person:
    def __init__(self):
        self._x = 10 #_xprotected , __x = private(can't be accessed in child class)

    def display(self):
        print(f"parent method, x= {self._x}")

class person2(person):
    pass

class student(person):
    def __init__(self):
        super().__init__() #for accessing parent variables

    def display1(self):
        print(f"protected {self. _x}") #itself bcs super().__init__(), apna hogya super().__init__() ki wajah se
        self.display()

s = student()
s.display1()
```

```python
#math module

import math
print("pi:", math.pi)
print("e:", math.e)
# Basic functions
print("ceil(4.2):", math.ceil(4.2)) # nearest upper integer
print("floor(4.8):", math.floor(4.8)) # nearest lower integer
print("fabs(-7.5):", math.fabs(-7.5)) # absolute float
print("factorial(5):", math.factorial(5)) # 5! = 120
print("sqrt(16):", math.sqrt(16)) # square root
print("pow(2,3):", math.pow(2,3)) # 2^3 = 8.0
print("exp(2):", math.exp(2)) # e^2
print("log(8,2):", math.log(8,2)) # log base 2
print("log10(100):", math.log10(100)) # log base 10
print("log2(16):", math.log2(16)) # log base 2
print("isnan(math.nan):", math.isnan(math.nan)) # check NaN
print("isinf(math.inf):", math.isinf(math.inf)) # check infinity
# Trigonometric functions
print("sin(pi/2):", math.sin(math.pi/2))
print("cos(0):", math.cos(0))
print("tan(pi/4):", math.tan(math.pi/4))
print("asin(1):", math.asin(1)) # inverse sine
print("acos(1):", math.acos(1)) # inverse cosine
print("atan(1):", math.atan(1)) # inverse tangent
print("atan2(1,1):", math.atan2(1,1)) # atan(y/x)
print("degrees(pi):", math.degrees(math.pi))
print("radians(180):", math.radians(180))
# Other useful constants / functions
print("copysign(5,-1):", math.copysign(5,-1)) # 5 with sign of -1
print("fmod(10,3):", math.fmod(10,3)) # remainder like %
print("gcd(12,18):", math.gcd(12,18)) # greatest common divisor
print("lcm(12,18):", math.lcm(12,18)) # least common multiple (Python 3.9+)
```
