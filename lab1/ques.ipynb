```python
def calculateSalary(basicSalary):
    hra = (0.1 * basicSalary)
    da = (0.05 * basicSalary)
    totalSalary = basicSalary + hra + da
    return totalSalary
    
basicSalary = int(input("enter salary: "))
result = calculateSalary(basicSalary)
print("total salary: ", result)
```

```python
def analyzeString(stringInput):
    result = {
        "total characters": 0,
        "uppercase letters": 0,
        "lowercase letters": 0,
        "digits": 0
    }

    for char in stringInput:
        result["total characters"] += 1
        if char.isupper():
            result["uppercase letters"] += 1 
        elif char.islower():
            result["lowercase letters"] += 1
        elif char.isdigit():
            result["digits"] += 1

    return result

string = input("enter a string: ")
output = analyzeString(string)

print("analysis: ")
for key, value in output.items():
    print(key, value)
```

```python
string = input("enter a string: ")

maxLength = 1
currentLength = 1
longestSubstring = string[0]
currentSubstring = string[0]

for i in range(1, len(string)):
    if string[i] == string[i-1]:
        currentLength += 1
        currentSubstring += string[i]
    else:
        currentLength = 1
        currentSubstring = string[i]

    if currentLength > maxLength:
        maxLength = currentLength
        longestSubstring = currentSubstring

print("lonest substring: ", longestSubstring)
print("length: ", maxLength)
```

```python
def findDuplicates(numbers):
    duplicates = []

    for i in range(len(numbers)):
        count = 0
        for j in range(len(numbers)):
            if numbers[i] == numbers[j]:
                count += 1

        if count > 1 and numbers[i] not in duplicates:
            duplicates.append(numbers[i])

    return duplicates

n = int(input("enter n: "))
nums = []

for i in range(n):
    nums.append(int(input("enter a number: ")))

result = findDuplicates(nums)
print("repeated numbers: ", result)
```
